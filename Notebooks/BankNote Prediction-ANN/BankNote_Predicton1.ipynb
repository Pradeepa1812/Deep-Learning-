**Predicting the BankNote is genunine or not using deep learning**

```
import tensorflow 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#read the dataset using pandas
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bank_note_data/training_set_label.csv" )
```

```
data.head()
```

```
data.shape
```

```
data.info()
```

```
#Splitting independent and dependend vaiable
X=data.iloc[:,0:4]
y=data.iloc[:,-1]
print(X)
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```
import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```
X_train.shape
```

```
#initializing
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=4))
model.add(Dense(1,activation='sigmoid'))
model.add(Dense(1))
```

```

model.summary()
```

```
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.fit(X_train,y_train,epochs=30,batch_size=20)
```

```
model.evaluate(X_test,y_test)
```

```
#Prediction
y_pred=model.predict(X_test)
```

```
import math
```

```
y_pred
```

```
y_prediction=[]
for y_predi in y_pred:
  if y_predi>0.5:
    y_predi=1
    y_prediction.append(y_predi)
  else:
    y_predi=0
    y_prediction.append(y_predi)
```

```
y_pred=[]
for i in y_prediction:
  y_pred.append(i)
y_pred
```

```
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_prediction))
```
